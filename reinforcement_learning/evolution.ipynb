```python
import cma
import numpy as np
```

```python
es = cma.CMAEvolutionStrategy(x0=np.zeros(8), sigma0=0.5)

while not es.stop():
    solutions = es.ask()
    es.tell(solutions, [cma.ff.rosen(x) for x in solutions])
    es.logger.add()  # write data to disc to be plotted
    es.disp()
```

```python
es.result_pretty()
cma.plot()
```
