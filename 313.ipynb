```python
n = 1000000
XYs = []
for i in range(n):
    (X_try, Y_try) = np.random.uniform(0, 1, size=2)
    if np.random.uniform(0, 1) < X_try + Y_try:
        XYs.append((X_try, Y_try))

Zs = []
for XY in XYs:
    X, Y = XY
    Zs.append(np.min((X, 2*Y)))
```

```python
res = 100
support = np.linspace(0, 1, res)
fz = np.zeros(res)
for i in range(len(support)-1):
    ds = support[i+1] - support[i]
    if Zs[i] > support[i] and Zs[i] < support[i+1]:
        fz[i] += 1
```

```python
Zs = np.array(Zs)
plt.figure()
plt.hist(Zs, bins=support, density=True, cumulative=False)
plt.plot(support, 1/8*(-9*support**2+10*support+6))
plt.show()
```
