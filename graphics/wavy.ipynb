```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import interp
from perlin_noise import PerlinNoise
%matplotlib widget

# Create a grid of points
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

# # Generate wavy block colors
# Z = np.sin(X) + np.cos(Y)

# # Generate a randomized wave
# Z = np.random.rand(100, 100)
# # Generate a smoother wave
Z = np.sin(X) + np.cos(Y) + np.random.rand(100, 100) * 0.1

# Generate a natural looking wave using Perlin noise


# noise = PerlinNoise(octaves=6, seed=1)
# Z = np.array([[interp(noise([i/10, j/10]), [-1, 1], [0, 1]) for j in range(100)] for i in range(100)])


# Generate a smoother Perlin noise wave with larger amplitudes
noise = PerlinNoise(octaves=8, seed=1)
Z = np.array([[interp(noise([i/20, j/10]), [-1, 1], [0, 1]) for j in range(100)] for i in range(100)])




# Plot the wavy block colors
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='GnBu', levels=5)
plt.colorbar(label='Intensity')
plt.title('Wavy Block Colors')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
