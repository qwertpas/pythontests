```python
import math

def cordic_sine(angle_rad, iterations=16):
    # Precompute arctangents and scaling factors for iterations
    angles = [math.atan(2**(-i)) for i in range(iterations)]
    scaling_factors = [1 / math.sqrt(1 + 2**(-2*i)) for i in range(iterations)]
    
    # Initialize initial vector and current angle
    x = 1.0
    y = 0.0
    z = angle_rad
    
    # Iteratively rotate the vector to approach the desired angle
    for i in range(iterations):
        if z < 0:
            x, y, z = x - y * 2**(-i), y + x * 2**(-i), z + angles[i]
        else:
            x, y, z = x + y * 2**(-i), y - x * 2**(-i), z - angles[i]
    
    # Multiply by the final scaling factor to get the sine value
    return y * scaling_factors[-1]

if __name__ == "__main__":
    # Example usage
    angle_deg = 45
    angle_rad = math.radians(angle_deg)
    
    # Calculate sine using CORDIC
    sine_cordic = cordic_sine(angle_rad)
    
    # Compare with the actual sine value from Python's math library
    sine_actual = math.sin(angle_rad)
    
    print(f"Angle (degrees): {angle_deg}")
    print(f"CORDIC Sine: {sine_cordic:.6f}")
    print(f"Actual Sine: {sine_actual:.6f}")
```
